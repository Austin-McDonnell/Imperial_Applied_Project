```python
def long_lat_price_scatter(data, plot_name):
    trace = go.Scatter(
        x = data.Latitude,
        y = data.Longitude,
        name = plot_name,
        mode='markers',
        opacity = 0.7,
        marker=dict(
            size=6,
            showscale=False,
            colorscale='RdYlBu',#'Blues',
            #reversescale=True,
            color=data.Price,
            #color = '#2ca02c', #set color equal to a variable
        )
    )
    return trace
```

```python
from plotly.subplots import make_subplots
price_plot_titles = [
    '$0 - $20k',
    '$20k - $30k',
    '$30k - $40k',
    '$40k - $60k',
    '$60k - $100k',
    '$100k - $180k',
    '$180k - $300k',
    '$300k+',
]

fig = make_subplots(rows=4, cols=2)#, shared_yaxes=True, shared_xaxes=True)

row_val=1
col_val=1
for bucket in price_plot_titles:
    #print(bucket)
    if row_val > 4:
        col_val+=1
        row_val=1
    temp = data[data.Price_Bin == bucket]
    fig.add_trace(long_lat_price_scatter(temp, bucket), row=row_val, col=col_val)
    row_val+=1
```
